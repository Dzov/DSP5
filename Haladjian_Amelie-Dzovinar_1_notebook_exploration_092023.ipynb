# Importing Data

We are given access to the data of a Brazilian ecommerce. 
The data is divided into 9 sets and contains the information of 100k orders from 2016 to 2018. 

```python
raw_customers = pd.read_csv("data/olist_customers_dataset.csv")
raw_items = pd.read_csv("data/olist_order_items_dataset.csv")
raw_payment = pd.read_csv("data/olist_order_payments_dataset.csv")
raw_reviews = pd.read_csv("data/olist_order_reviews_dataset.csv")
raw_orders = pd.read_csv("data/olist_orders_dataset.csv")
raw_products = pd.read_csv("data/olist_products_dataset.csv")
raw_sellers = pd.read_csv("data/olist_sellers_dataset.csv")
raw_geoloc = pd.read_csv("data/olist_geolocation_dataset.csv")
raw_category_translation = pd.read_csv(
    "data/product_category_name_translation.csv")

data_sets = {
    "Customers": raw_customers,
    "Orders": raw_orders,
    "Order Items": raw_items,
    "Order Reviews": raw_reviews,
    "Order Payment": raw_payment,
    "Products": raw_products,
    "Sellers": raw_sellers,
    "Geolocation": raw_geoloc,
    "Category translation": raw_category_translation,
}
```

# Analysis

```python
def plot_missing_data(df, name):
    plt.figure(figsize=[15, 3])
    df.isna().mean().plot(kind="bar")
    plt.title("Missing values by feature")
    plt.ylabel("Proportion of Missing Values")
    plt.xticks(rotation=25, ha="right")
    plt.show()


def describe(df):
    summary = df.describe(include="all")
    summary.loc["nan_count"] = df.isna().sum()
    display(summary)


for name, set in data_sets.items():
    null_percentage = set.isna().mean().mean() * 100
    display(Markdown(f"## {name}"))
    display(
        Markdown(
            f" Shape: {str(set.shape)}, Null Values: {str(round(null_percentage, 1))}%"
        )
    )
    display(set.sample(2))
    describe(set)
    if null_percentage > 0:
        plot_missing_data(set, name)
    print("\n")
```

Olist provided a clean data set, with little missing values. The only actual missing data is related to product description, which will not be used for further analysis. 

The missing values concerning reviews will be imputed with placeholders indicating that the customer did not leave a review comment.

## Merging data sets

#### Orders

We can start by merging the customer data set with their orders

```python
def merge(df1, df2, key):
    return pd.merge(df1, df2, on=key, how="left")
```

```python
orders = merge(orders, raw_customers, "customer_id")
```

```python
describe(orders)
```

There are 8 orders for which we do not have the delivery date 

```python
orders.loc[orders["order_delivered_customer_date"].isna()]
```

We can check in the reviews if we can determine whether they did receive their orders 

```python
orders_ids = orders.loc[orders["order_delivered_customer_date"].isna()]["order_id"]

reviews_subset = raw_reviews.loc[raw_reviews["order_id"].isin(orders_ids)]

reviews_subset
```

According to the reviews, they all received their orders. We will therefore impute the missing delivery dates with the review dates 

```python
mapping_dict = dict(
    zip(reviews_subset["order_id"], reviews_subset["review_creation_date"])
)

orders.loc[
    orders["order_id"].isin(orders_ids), "order_delivered_customer_date"
] = orders["order_id"].map(mapping_dict)

orders.loc[orders["order_id"].isin(orders_ids)]
```

We can also remove the features that we will not be needing going forward 

```python
orders.drop(
    columns=[
        "order_approved_at",
        "order_delivered_carrier_date",
        "customer_zip_code_prefix",
    ],
    inplace=True,
)
```

#### Products 

```python
merged_df = merge(raw_products, raw_category_translation, "product_category_name")
raw_products["product_category_name"] = merged_df["product_category_name_english"]
```

```python
raw_products
```

#### Items 

```python
raw_items = merge(raw_items, raw_products, "product_id")
```

```python
raw_items
```

We can aggregate the items data set to only have one line per order with the total price and freight for the entire order. 

This will allow us to have a wider picture of a customer's order.

```python
def count_unique_occurences(series):
    return len(series.unique())


def list_values(series):
    return series.unique()
```

```python
raw_items["total_price"] = raw_items["price"] + raw_items["freight_value"]

items = (
    raw_items.groupby("order_id")
    .agg(
        total_price=("total_price", "sum"),
        item_count=("order_item_id", "count"),
        category_count=("product_category_name", count_unique_occurences),
        categories=("product_category_name", list_values),
    )
    .reset_index()
)
```

```python
items.loc[items.category_count > 1]
```

We can now merge these data sets to have only one set that contains all the information regarding a customer's orders 

```python
orders = merge(orders, items, "order_id")
```

```python
summary = orders.describe(include="all")
summary.loc["nan_count"] = orders.isna().sum()
summary
```

#### Reviews

Some orders have multiple reviews, we will aggregate the values and then merge reviews into our order dataframe.

```python
raw_reviews["comment"] = (
    (raw_reviews["review_comment_title"].notna())
    | (raw_reviews["review_comment_message"].notna())
).astype(int)

reviews = (
    raw_reviews.groupby("order_id")
    .agg({"review_score": "mean", "comment": "max"})
    .reset_index()
)

reviews
```

```python
orders = merge(orders, reviews, "order_id")
```

```python
describe(orders)
```

There are 646 orders without reviews. We will impute these missing scores with a 3, which is neither a good nor a bad review. 

```python
orders["review_score"].fillna(3, inplace=True)
orders["comment"].fillna(0, inplace=True)
```

```python
payments = (
    raw_payment.groupby("order_id")
    .agg(
        payment_installments=("payment_installments", "sum"),
        mean_installment_payment_value=("payment_value", "mean"),
        payment_type_count=("payment_type", count_unique_occurences),
    )
    .reset_index()
)

payments
```

```python
orders = merge(orders, payments, "order_id")
```

```python
describe(orders)
```

There is one order without payment information

```python
orders.loc[
    (orders["mean_installment_payment_value"].isna())
    | orders["payment_installments"].isna()
]
```

We will impute the values with default values 

```python
orders["mean_installment_payment_value"].fillna(orders["total_price"], inplace=True)
orders["payment_installments"].fillna(1, inplace=True)
orders["payment_type_count"].fillna(1, inplace=True)
```

There are orders with 0 payment installments, but the default should be one 

```python
orders.loc[orders["payment_installments"] == 0]
```

```python
orders.loc[orders["payment_installments"] == 0, "payment_installments"] = 1
```

We now have a comprehensive data set with the order information for all the customers. 

We can now proceed to the feature engineering.


# Feature Engineering

```python
describe(orders)
```

We can create a feature to indicate whether a customer received their order on time 

```python
date_features = [
    "order_purchase_timestamp",
    "order_delivered_customer_date",
    "order_estimated_delivery_date",
]

orders[date_features] = orders[date_features].apply(pd.to_datetime)
```

```python
orders["late_delivery"] = (
    orders.order_delivered_customer_date > orders.order_estimated_delivery_date
).astype(int)
```

Then add a feature to indicate purchase recency, based on the latest purchase date from this data set 

```python
latest_purchase = pd.to_datetime(orders["order_purchase_timestamp"].max())

latest_purchase
```

```python
orders["order_purchase_timestamp"][1]
```

```python
orders["recency"] = (
    latest_purchase - orders["order_purchase_timestamp"]).dt.days
```

```python
orders["recency"]
```

```python
orders.loc[orders.order_delivered_customer_date >
           orders.order_estimated_delivery_date]
```

We can now aggregate the data to create an average order for returning customers 

```python
returning_customers = raw_customers.loc[
    raw_customers["customer_unique_id"].duplicated()
]
returning_customers.shape
```

```python
customer_summary = (
    orders.groupby("customer_unique_id")
    .agg(
        order_frequency=("order_id", "count"),
        total_price=("total_price", "sum"),
        mean_item_count=("item_count", "mean"),
        mean_category_count=("category_count", "mean"),
        review_score=("review_score", "mean"),
        comment=("comment", stats.mode),
        payment_installments=("payment_installments", "mean"),
        payment_type_count=("payment_type_count", stats.mode),
        late_delivery=("late_delivery", stats.mode),
        recency=("recency", "min"),
    )
    .reset_index()
)

customer_summary
```

We now have a dataframe that contains the average cart of each unique customer

```python
customer_summary.to_csv("customers.csv")
```
